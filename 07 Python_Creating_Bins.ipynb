```python
import pandas as pd
cardio = pd.read_csv("cardio.csv")
cardio.head()
```

```python
# encode the categories for male and female
one_hot = pd.get_dummies(cardio, columns=["gender"])   
one_hot.head()
```

```python
# convert females as 1
cardio.loc[cardio.gender == "female", 'gender'] = 1
cardio
```

```python
# convert males as 0
cardio.loc[cardio.gender == "male", 'gender'] = 0
cardio
```

```python
cardio['weight'] = cardio.weight * 2.2046
```

```python
cardio.head()
```

```python
cardio.weight.describe()
```

```python
bins = range(100,350,50)
```

```python
[x for x in pd.cut(cardio.weight, bins).unique()]
```

```python
cardio["weight_bins"] = pd.cut(cardio.weight, bins=bins)
cardio
```

```python
cardio[["weight", "weight_bins"]].head()
```

```python
cardio.weight_bins.unique()
```

```python
labels = ["100-150", "150-200", "200-250", "250-300"]
```

```python
cardio["weight_bins"] = pd.cut(cardio.weight, 
                              bins=bins,
                              labels=labels)
```

```python
cardio[["weight", "weight_bins"]][cardio.weight_bins.isnull()]
```

```python
cardio = cardio.dropna(axis=0, 
                      subset=["weight_bins"])
```
